# Write multiple lines in a file

```python
with open("greeting.txt", "w") as file:
    file.write("Namaste, welcome to Python \n")
    file.write("this is the second line")
```

```python
with open("greeting.txt", "r") as file:
    print(file.read())
```

```python
# Create a sample file for our reading experiments
with open("students.txt", "w") as file:
    file.write("Rahul Sharma - Mumbai - 85\n")
    file.write("Priya Patel - Delhi - 92\n")
    file.write("Amit Kumar - Bangalore - 78\n")
    file.write("Sneha Reddy - Hyderabad - 95\n")
    file.write("Vikram Singh - Pune - 88\n")

print("✅ students.txt created!")
```

```python
with open("students.txt", "r") as file:
    print(file.read())
    print(type(file.read()))
```

```python
# You can see the \n characters if you use repr()
with open("students.txt", "r") as file:
    content = file.read()

print(repr(content))
```

```python
# Create a sample file for our reading experiments
with open("students_new.txt", "w") as file:
    file.write("Rahul Sharma - Mumbai - 85\n Priya Patel - Delhi - 92\n Amit Kumar - Bangalore - 78\n")

print("✅ students.txt created!")
```

```python
citites = ["Mumbai \n", "Kakinada \n", "Pune \n", "Chennai \n",]

with open("students_new.txt", "w") as file:
    file.writelines(citites)



```
